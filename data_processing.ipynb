```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import textblob
from textblob import Word
import string
import re
nltk.download('all')
```

```python
product = pd.read_csv('product_caption.csv')
```

```python
product
```

```python
new_product_caption_list = list()
for row in product['caption']:
    row = re.sub(r'\[Music]', '', row)
    row = row.translate(str.maketrans('','',string.punctuation))
    new_product_caption_list.append(row.lower())   
product['caption'] = new_product_caption_list
```
